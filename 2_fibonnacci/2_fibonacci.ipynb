### fibonnaci en python

cantidad de iteraciones para hallar el n-esimo termino de la sucesion de fibonacci

importar los modulos necesarios y declaracion de variables

```python
import matplotlib.pyplot as plt
import time

n= 1000  # n-esimo termino
x= 1
anterior = 0;
i=1;
start_time = time.time()
fibonacci_time=[]
```

Algoritmo para hallar la serie de fibonacci

```python
while( i <= n ):
	temporal = x
	x = x + anterior
	anterior = temporal
	i=i+1
	fibonacci_time.append(time.time()-start_time)
```

Uso de la libreria mathplotlib para graficar el tiempo para hallar el e-nesimo valor de la serie de fibonacci 

```python
plt.plot(fibonacci_time,label="tiempo de ejecucion")    
plt.title("fibonacci")   
plt.xlabel("N-esimo termino")   
plt.ylabel("Tiempo (s)")
plt.legend()
plt.show()
```
