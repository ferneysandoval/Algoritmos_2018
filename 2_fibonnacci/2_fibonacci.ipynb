## tiempo de ejecuccion del n-esimo termino de la serie de fibonacci

cantidad de iteraciones para hallar el n-esimo termino de la sucesion de fibonacci

importar los modulos necesarios y declaracion de variables.


```python
import matplotlib.pyplot as plt
import numpy as np
import time
```

Algoritmo para hallar la serie de fibonacci


```python
def fibo(n):
    start_time=time.time()
    a=0
    b=1
    for i in range(n):
        a,b =b,a+b
    return time.time()-start_time
```



Uso de la libreria mathplotlib para graficar el tiempo para hallar el e-nesimo valor de la serie de fibonacci


```python
ns=[]
ns= np.arange(1,50000,100)
fibonnacci =[]
for n in ns:
    fibonnacci.append(fibo(n))
plt.plot(ns,fibonnacci)
plt.xlabel("n-esimo termino")
plt.ylabel("tiempo(s)")
```
